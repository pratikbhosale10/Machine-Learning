```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
import pandas as pd

df = pd.DataFrame(data = cancer['data'], columns = cancer['feature_names'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['target'] = cancer['target']
```

```python
df.head()
```

```python
df.corr()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
scaled_array = ss.fit_transform(df.drop('target', axis = 1 ))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(df.corr(), cmap = 'coolwarm')
```

```python
#PCA is used to merge multiple columns. In above dataset total 30 columns are there for X. It will create confusion for machine while training and prediction. In order to avoid this PCA is used.

from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2) # n_components is used to to give number of columns we want.
```

```python
x_pca = pca.fit_transform(scaled_array)
```

```python
x_pca.shape
```

```python
x_pca
```

```python
y = df['target']
```

```python
y
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1], c = y)  # we have taken first column of x_pca as X and 2nd column of x_pca as Y. c is used to give color in scatter plot.
```

```python
# from above plot we are not able to understand which dots are for 0 and which are for 1.For that we are using conditions y==0 and y==1

plt.scatter(x_pca[y==0,0], x_pca[y==0,1], color = 'red')    # it will select all columns of x_pca and will select only those rows which satisfies y==0 condition.
plt.scatter(x_pca[y==1,0], x_pca[y==1,1], color = 'yellow')  # it will select all columns of x_pca and will select only those rows which satisfies y==1 condition.
```
